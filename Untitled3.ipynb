Import dataset

```python
import pandas as pd
superstore = pd.read_csv('superstore_dataset2011-2015.csv', parse_dates=True, encoding="ISO-8859-1")
print('Dataset size: %d rows and %d columns\n' % superstore.shape)
print('First five data: ')
superstore.head()
```

This is a retail dataset of a global superstore for four years

```python
superstore.dtypes
```

Add new column Order Month from column Order Date

```python
import datetime
def convert(x):
    date_format = "%d/%m/%Y" if '/' in x else "%d-%m-%Y"
    return datetime.datetime.strptime(x, date_format).strftime('%m/%Y')
    
superstore['Order Month'] = superstore['Order Date'].apply(convert)
superstore['Order Month'] = superstore['Order Date'].apply(lambda x: datetime.datetime.strptime(x, "%d/%m/%Y" if '/' in x else "%d-%m-%Y").strftime('%m/%Y'))
superstore.head()
```
